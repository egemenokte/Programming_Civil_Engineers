Let's do an example, let's write a function that returs the factorial of a number! We need a loop!

````python
def factorial(num):
    '''
    returns the factorial given num
    num must be an integer
    '''
    fact=1
    for i in range(1,num+1):
        fact=fact*i


    return fact
````

````python
factorial(3)
````

````python
#Let's make sure the user passes an integer and raise a warning if they do not

def factorial(num):
    '''
    returns the factorial given num
    num must be an integer
    '''

    if type(num) != int: #this is called error handling. Can get more complicated
        raise Exception('This is not an integer!')

    fact=1
    for i in range(1,num+1):
        fact=fact*i


    return fact
````

###Local and Global Variables
A variable declared inside a function is called local variable, while a variable not bound to any function but can be accessed outside the function is called global variable.

````python
#Let's give an example from Weight

def Weight(m):
    g=9.81 #m/2
    return m*g #in Newtons
````

````python
Weight(10)
````

````python
g #g is not defined! G is a local variable, only existing inside the function
````

For the function to access g, I must either declare it global or define it outside the function

````python
#Let's give an example from Weight

def Weight(m):
    global g #define g as a global variable
    g=9.81 #m/2
    return m*g #in Newtons
````

````python
print(Weight(10))
g #now g prints no problem!
````

I can also declare it outside of the function!

````python
#Let's give an example from Weight
g=9.81
def Weight(m):
    return m*g #in Newtons
````

````python
print(Weight(10)) #this still works!
````

## **Anonymous Functions (Lambda Functions)**
Anonymous functions, also known as lambda functions, are defined using the `lambda` keyword. They are called anonymous because they do not need to be named. They are useful for creating small, one-off functions that are often used where function objects are required, such as in arguments for higher-order functions. Here is an example:

````python
def square(a):
    y=a**2
    return y

print(square(10))
````

````python
#There is a much more compact way of doing this

square = lambda x : x**2 #map x to x**2
square(10)
````

## Classes

### Introduction to Objects

In programming, objects are key elements of object-oriented programming (OOP). They encapsulate both data and the functions (methods) that operate on this data, combining them into cohesive units.

### Why Objects Are Useful and Different from Functions

1. **Encapsulation**: Objects keep data and methods that modify this data together, enhancing security and ease of maintenance.
2. **Reusability**: Objects can be reused across different parts of a program, saving time and reducing errors.
3. **Inheritance**: Objects can inherit characteristics from other objects, simplifying code reuse and maintenance.
4. **Polymorphism**: Objects can process data differently depending on their subtype, which adds flexibility to the code.

### Differences from Functions

Unlike functions, which are procedures that do not inherently encapsulate or hold onto data, objects manage their own data and maintain state across method calls. This structure allows objects to interact more effectively, especially in large, complex systems, by mimicking real-world interactions.

````python
class RoadSection:
    """Represents a single section of a road."""

    def __init__(self, name, length, speed_limit):
        self.name = name
        self.length = length  # In miles
        self.speed_limit = speed_limit  # In mi/h

    def calculate_travel_time(self):
        """Calculates the travel time in seconds."""
        time_in_min = self.length / (self.speed_limit)  # Convert length to km
        return time_in_min * 60  # Convert hours to seconds
````

````python
Amherst_to_Hadley=RoadSection(name='Route_9',length=6,speed_limit=45)
print(Amherst_to_Hadley.length)
print(Amherst_to_Hadley.calculate_travel_time())
````

````python
class RoadNetwork:
    """Represents a network of connected road sections."""

    def __init__(self):
        self.sections = []

    def add_section(self, section):
        self.sections.append(section)

    def calculate_total_travel_time(self):
        """Calculates the time to traverse the entire network in hours."""
        total_time = 0
        for section in self.sections:
            total_time += section.calculate_travel_time()
        return total_time
````

````python
Amherst_to_Hadley=RoadSection(name='Route 9',length=6,speed_limit=55)
Hadley_to_Northampton=RoadSection(name='Route 9',length=6,speed_limit=45)
Northampton_to_Springfield=RoadSection(name='I91',length=27,speed_limit=65)
````

````python
Network = RoadNetwork()
Network.add_section(Amherst_to_Hadley)
Network.add_section(Hadley_to_Northampton)
Network.add_section(Northampton_to_Springfield)
print(Network.calculate_total_travel_time())
````

### Babylonian Method for Finding Square Roots

The Babylonian method, also known as Heron's method, is an ancient algorithm for approximating the square root of a positive number. The method iteratively improves an initial guess through a simple averaging process. Here's a step-by-step explanation of the method:

#### Step-by-Step Algorithm

1. **Initial Guess:**
   - Start with an initial guess for the square root. A common choice is half of the number whose square root is being calculated.

2. **Iterative Improvement:**
   - Update the guess by averaging it with the number divided by the guess.

3. **Stopping Criteria:**
   - Repeat the iteration until the desired accuracy is achieved. This can be controlled by a maximum number of iterations or a tolerance level that stops the process when the change in the guess becomes very small.

#### Mathematical Formula

Given a number $ x $ for which we want to find the square root, the Babylonian method iteratively applies the following formula:

$ \text{new_guess} = \frac{(\text{old_guess} + \frac{x}{\text{old_guess}})}{2} $

#### Pseudo Algorithm

1. Initialize `guess` to $ \frac{x}{2} $
2. Loop for a fixed number of iterations or until the change in `guess` is below a specified tolerance:
   - Calculate `new_guess` as the average of `guess` and $ \frac{x}{\text{guess}} $.
   - If the absolute difference between `new_guess` and `guess` is less than the tolerance (say $10^{-3}$), break the loop.
   - Update `guess` to `new_guess`.
3. Return the final `guess` as the approximated square root.
